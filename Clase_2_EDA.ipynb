# Exploremos matrices de datos con NumPy:

Comencemos con unos datos sencillos.

Supongamos que un profesor toma una muestra de las calificaciones de sus estudiantes para analizarlas (estas van de 0 a 100):

```python
import numpy as np
```

```python
data = [50,50,47,97,49,3,53,42,26,74,82,62,37,15,70,27,36,35,48,52,63,64]
print(data)
```

Los datos se cargan en Python en formato tipo **lista**, la cual es buena para manipulación de información pero no tanto para análisis numéricos. Por lo tanto vamos a usar el paquete **NumPy** que importamos en la línea 2, el cual incluye tipos específicos de datos y funciones para trabajar con _números_ en Python:

```python
notas = np.array(data)
print(notas)
```

Miremos la diferencia entre una **lista** y un **array** (**matriz**), comparemos como se comportan cuando hacemos operaciones aritméticas con ella, como multiplicarlas por 2.

```python
print (type(data),'x 2:', data * 2)

print('---')

print (type(notas),'x 2:', notas * 2)
```

Nótese que para la lista se crea una nueva lista con el doble de elementos de la secuencia original repetidos. Mientras que la matriz o array de Numpy multiplica cada elemento por 2, tratándolo entonces como un _vector_. Por lo tanto, NumPy nos permite realizar operaciones matemáticas sobre datos numéticos, lo que lo hace más útil para el análisis de datos que la lista genérica.

¿Qué significa **numpy.ndarray**?: El **nd** inica que es una estructura que puede consistir de múltiple dimensione (n-dimensiones). En nuestro caso, tiene una sola dimensión: las notas de los estudiantes. Miremos la forma de la matriz:

```python
notas.shape
```

La forma entonces consiste en una dimensión que contiene 22 eelementos. Podemos acceder a elementos de la matriz de la siguiente forma:

**Tener en cuenta que en Python el conteo de elementos comienza en cero**:


```python
notas[0]
```

Miremos ahora que podemos hacer con los datos de calificaciones:

```python
notas.mean()
```

Entonces el promedio es aproximadamente 50, lo que es aproximadamente la mitad del rango posible de 0 a 100.

Añadamos ahora un segundo conjuntoi de datos para los estudiantes. Esta vez las horas semanales que dedicaron a estudiar:

```python
# Definir una matriz de horas de estudio
horas_estudio = [10.0,11.5,9.0,16.0,9.25,1.0,11.5,9.0,8.5,14.5,15.5,
               13.75,9.0,8.0,15.5,8.0,9.0,6.0,10.0,12.0,12.5,12.0]

# Crear una matriz 2D (una matriz de matrices)
datos_estudiantes = np.array([horas_estudio, notas])

# Mostrar la matriz
datos_estudiantes
```

La lista consiste entonces de una matriz en 2D. Miremos su forma:

```python
datos_estudiantes.shape
```

__datos_estudiantes__ contiene 2 elementos, cada uno una matriz que contiene 22 elementos. Para navegar en los datos, necesitamos definir ahora la posición de cada elemento en la jerarquía. Por ejemplo, el primer elemento de la primera matriz se llamaría de la siguiente forma:

```python
datos_estudiantes[0][0]
```

Ahora que tenemos una matriz de 2D que contiene las horas estudiadas y la calificación obtenida por lo estudiantes, podemos comparar ambas variables:

```python
prom_estudio = datos_estudiantes[0].mean()
prom_calificacion = datos_estudiantes[1].mean()

print('Promedio de horas de estudio: {:.2f}\nCalificación promedio: {:.2f}'.format(prom_estudio, prom_calificacion))
```

## Explorando _datos tabulares_ con __Pandas__:

Ya vimos que NumPy permite trabajar con números a través de matriz de valores numéricos. Sin embargo, cuando comenzamos a trabajar con tablas de datos de 2 dimensiones (filas y columnas), el paquete __Pandas__ ofrece una estructura más conveniente para realizar operaciones: el __DataFrame__:

Coarramos lo siguiente y miremos que hace:

```python
import pandas as pd

df_estudiantes = pd.DataFrame({'Nombre': ['Daniel', 'Johan', 'Pedro', 'Rosa', 'Elkin', 'Victoria', 'Federico', 'Jaime', 
                                     'Gloria', 'Giovanny', 'Francisca', 'Esteban', 'Daniela', 'Manuela', 'Jenny',
                                     'Luis','Helena','Ismael','Anita','Santiago','Daniel','Alicia'],
                            'Horas_Estudio':datos_estudiantes[0],
                            'Calificacion':datos_estudiantes[1]})
df_estudiantes
```

## Encontrar y filtrar datos en un DataFrame:

Esto podemos hacerlo con el método **loc**, de la siguiente manera:

```python
df_estudiantes.loc[5]
```

Podemos obtener un rango:

```python
df_estudiantes.loc[0:5]
```

Con el __iloc__ podemos encontrar filas basadas no en el índice sino en su posición ordinal en la base de datos:

```python
df_estudiantes.iloc[0:5]
```

**¿Notan la diferencia?**

_iloc_ ubica según el la posición y _loc_ según el índice. Podemos usar _iloc_ para ubicar, por ejemplo, los valores de las columnas 1 y 2 correspondientes a la fila 0 de la siguiente manera:

```python
df_estudiantes.iloc[0,[1,2]]
```

__¿Cómo se haría con _loc_?__

Dado que las columnas no tienen un índice explícito se usan los nombres para identificarlas:

```python
df_estudiantes.loc[0, 'Calificacion']
```

Veamos otro ejemplo:

```python
df_estudiantes.loc[df_estudiantes['Nombre']=='Alicia']
```

Pero ... no necesitamos el _loc_ para hacer esto. Miremos:

```python
df_estudiantes[df_estudiantes['Nombre']=='Alicia']
```

Con el __query__ del DataFrame también podemos hacerlo:

```python
df_estudiantes.query('Nombre=="Alicia"')
```

¿Qué indica esto? Que en Pandas podemos llegar a un resultado de varias formas. Esta es otra, por ejemplo:

```python
df_estudiantes[df_estudiantes.Nombre=='Alicia']
```

## Tratando con datos faltantes (missing):

Es uno de los problemas más frecuentes a los que nos enfrentamos en el análisis de datos: información incompleta o faltante. ¿Cómo sabemos que el Data Frame contiene datos faltantes?. De la siguiente manera: 

```python
df_estudiantes.isnull()
```

Esto por supuesto es ineficiente a medida que aumenta el tamaño de la base de datos. Entonces podemos aplicar:

```python
df_estudiantes.isnull().sum()
```

Ahora sabemos que hay 5 datos faltantes en __HorasEstudio__ y 5 en __Calificacion__. Vamos a verlos en contexto filtrando sólo aquellas filas con columnas (eje 1 del DataFrame) tipo Null.

```python
df_estudiantes[df_estudiantes.isnull().any(axis=1)]
```

Cuando recuperamos este subconjunto de la base de datos los valores faltantes nos aparecen como __Nan__: _not a number_.

¿Qué podemos hacer con estos datos? Una opción que se usa comunmente es la _imputación_. Por ejemplo, si el número de horas de estudio falta podemos asumir que el estudiante estudió el promedio de todos sus compañeros y reemplazar su valor faltante por este promedio. Esto lo hacemos de la siguiente manera (método __fillna__):

```python
df_estudiantes.HoraEstudio = df_estudiantes.HoraEstudio.fillna(df_estudiantes.HoraEstudio.mean())

df_estudiantes
```

Otra alternativa es asegurarnos de usar información para la que tenemos absoluta certeza. En este caso, podemos eliminar filas o columnas que contengan valores nulos (Null) usando el método __dropna__:

```python
df_estudiantes = df_estudiantes.dropna(axis=0, how='any')

df_estudiantes
```

## Exploremos los datos del DataFrame:

Ahora que limpiamos los dato faltantes, exploremos la base de datos analizando la media de las horas de estudio y las calificaciones.

```python
media_estudio = df_estudiantes['HoraEstudio'].mean()

media_calificacion = df_estudiantes['Calificacion'].mean()

print('Hora promedio de estudio semanal: {:.2f}\nCalificacion promedio: {:.2f}'.format(media_estudio, media_calificacion))
```

Miremos los estudiantes que estudiaron más tiempo que la media:

```python
df_estudiantes[df_estudiantes.HoraEstudio> media_estudio]
```

Nótese que el resultado es por si mismo un DataFrame, por lo que podemos trabajar con este como con cualquier Data Frame.

Miremos por ejemplo el promedio de aquellos que estudian más que el promedio de los estudiantes.

```python
df_estudiantes[df_estudiantes.HoraEstudio> media_estudio].Calificacion.mean()
```

Asumamos que se pasa con 3.0.

Podemos añadir una columna que nos indique si los estudiantes aprobaron o no.

Primero, creamos una __Serie__ de _Pandas_ que nos de el indicador de aprobó/reprobó (True o False), después lo uniremos a una nueva columna del DataFrame original:

```python
aprobar = pd.Series(df_estudiantes['Calificacion'] >= 3)
df_estudiantes = pd.concat([df_estudiantes, aprobar.rename("Aprobo")], axis=1)

df_estudiantes
```

Los DataFrames están diseñados para datos tabulares y pueden usarse para realizar muchas operaciones de este tipo, como agrupar o agregar tablas de información.

Por ejemplo, podemos usar el método __groupby__ para agrupar a los estudiantes en grupos de acuerdo con la columna _Aprobo_ que creamos previamente. Es decir, podemos determinar cuantos aprobaron y cuantos reprobaron.

```python
print(df_estudiantes.groupby(df_estudiantes.Aprobo).Nombre.count())
```

Podemos agregar varios campos en un grupo. Por ejemplo, podemos encontrar el promedio de horas de estudio y calificación, para los estudiantes que aprobaron y los que reprobaron el curso:

```python
print(df_estudiantes.groupby(df_estudiantes.Aprobo)[['HoraEstudio', 'Calificacion']].mean())
```

Los DataFrame son muy versátiles y permiten manipular datos de forma sencilla. Si queremos conservar los cambios realizados sobre el DataFrame tenemos asignar la operación realizada a la variable original. Por ejemplo, acá los organizamos de forma descendente según su calificación y asignamos el resultado a la variable _df_estudiantes_:

```python
df_estudiantes = df_estudiantes.sort_values('Calificacion', ascending=False)

df_estudiantes
```

El fundamento de la ciencia de datos es la __estadística__, aprendamos entonces algunas técnicas básicas:

## Estadística descriptiva y distribución de los datos:

Cuando examinamos una _variable_, por ejemplo las calificaciones, como científico de datos podemos interesarnos en su _distribución_, en otras palabras, como los diferentes valores para las calificaciones se distribuyen en la muestra. El punto de partida para esta exploración usualmente consiste en visualizar los datos como un histograma y mirar con cuanta frecuencia ocurre cada valor.

```python
var_datos = df_estudiantes['Calificacion']

fig = plt.figure(figsize=(10,4))

plt.hist(var_datos)

plt.title('Distribución de los datos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.show()
```

El histograma de las calificaciones tiene un aparente sesgo a la derecha, con los valores más frecuentes ocurriendo entre 3.5 y 4. Con algunos extremos concentrados en valores muy bajos.

## Medidas de tendencia central:

Para entender mejor la distribución, podemos examinar las _medidas de tendencia central_, lo cual representa la mitad de los datos y permite encontrar los valores "típicos". Algunas formas de definir la mitad de los datos son:

- El _promedio_: un promedio simple basado en la suma de todos los valores dividido por el número de observaciones.
- La _media_: el valor en la mitad de todos los valores de la observación.
- La _moda_: el valor de más común ocurrencia en la muestra.

Calculemos estos valores así como el _mínimo_ y el _máximo_ para realizar comparaciones, y mostremos estos en el histograma:

```python
var = df_estudiantes['Calificacion']

min_val=var.min()
max_val=var.max()
prom_val=var.mean()
media_val=var.median()
moda_val= var.mode()[0]

print('Mínimo:{:.2f}\nPromedio:{:.2f}\nMedia:{:.2f}\nModa:{:.2f}\nMáximo:{:.2f}\n'.format(min_val,
                                                                                        prom_val,
                                                                                        media_val,
                                                                                        moda_val,
                                                                                        max_val))

fig = plt.figure(figsize=(10,4))

plt.hist(var)

plt.axvline(x=min_val, color = 'gray', linestyle='dashed', linewidth = 2)
plt.axvline(x=prom_val, color = 'cyan', linestyle='dashed', linewidth = 2)
plt.axvline(x=media_val, color = 'red', linestyle='dashed', linewidth = 2)
plt.axvline(x=moda_val, color = 'yellow', linestyle='dashed', linewidth = 2)
plt.axvline(x=max_val, color = 'gray', linestyle='dashed', linewidth = 2)


plt.title('Distribución de los datos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.show()
```

Otro gráfico que permite visualizar la distribución de los datos es el _box plot_ a veces llamado diagrama de bigotes o diagrama de caja. Miremos:

```python
var = df_estudiantes['Calificacion']

fig = plt.figure(figsize=(10,4))

plt.boxplot(var)

plt.title('Distribución de los datos')

plt.show()
```

El _box plot_ muestra la distribución de los datos de la siguiente forma: la caja muestra donde residen los dos cuartiles interiores de los datos. La mitad de las notas están aproximadamente entre 2.8 y 4.0. Los bigotes, que se extienden desde los bordes de las cajas muestran donde se encuentran los otros dos cuartiles. En este caso, aproximadamente desde 2.1 hasta 4.9. La línea naranja indica la media. El punto que aparece en 1 indica un valor atípico.

A veces es útil combinar histogramas y boxplot:

```python

def show_distribution(var_data):
    from matplotlib import pyplot as plt

    min_val = var_data.min()
    max_val = var_data.max()
    mean_val = var_data.mean()
    med_val = var_data.median()
    mod_val = var_data.mode()[0]

    print('Mínimo:{:.2f}\nPromedio:{:.2f}\nMedia:{:.2f}\nModa:{:.2f}\nMáximo:{:.2f}\n'.format(min_val,
                                                                                            mean_val,
                                                                                            med_val,
                                                                                            mod_val,
                                                                                            max_val))

   
    fig, ax = plt.subplots(2, 1, figsize = (10,4))

     
    ax[0].hist(var_data)
    ax[0].set_ylabel('Frecuencia')

   
    ax[0].axvline(x=min_val, color = 'gray', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mean_val, color = 'cyan', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=med_val, color = 'red', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mod_val, color = 'yellow', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=max_val, color = 'gray', linestyle='dashed', linewidth = 2)

      
    ax[1].boxplot(var_data, vert=False)
    ax[1].set_xlabel('Valor')

    
    fig.suptitle('Distribución de los datos')

   
    plt.show()


    
col = df_estudiantes['Calificacion']
   
show_distribution(col)
```

En el output anterior podemos que todas las medidas de tendencia central están a la derecha de la distribución de los datos. Con los menores valores a su izquierda. 

Para explorar esta distribución con mayor detalle debemos entender que la estadística es principalmente tomar muestras de datos y usar funciones de probabilida para extrapolar información acerca de la población total de los datos.

¿Que significa esto? _muestra_ se refiere a los datos que tenemos disponibles, en este caso las calificación de los estudiantes y sus habitos de estudio. La _poblacilón_ se refiere a todos los datos que podríamos recolectar, tales como cada calificación de los estudiantes y los hábitos de estudio en todas las escuelas en toda la historia de la humanidad. Casi siempre estamos interesados en la población, pero no es práctico ni posible recoger toda esta información. En cambio, podemos estimar como es la población a partir de pequeños conjuntos de datos (muestra) disponibles.

Si tenemos suficiente muestras podemos caluclar algo llamado _función de densidad de probabilidad_ la cual estima la distribución de las calificaciones en toda la población.

La clase __pyplot__ de Matplotlib proporciona una función de gráfico que es útil para ver la densidad:

```python
def show_density(var_datos):
    from matplotlib import pyplot as plt

    fig = plt.figure(figsize=(10,4))

    
    var_datos.plot.density()


    plt.title('Densidad de los datos')

   
    plt.axvline(x=var_datos.mean(), color = 'cyan', linestyle='dashed', linewidth = 2)
    plt.axvline(x=var_datos.median(), color = 'red', linestyle='dashed', linewidth = 2)
    plt.axvline(x=var_datos.mode()[0], color = 'yellow', linestyle='dashed', linewidth = 2)

 
    plt.show()


col = df_estudiantes['Calificacion']
show_density(col)
```

Como se esperaba según el histograma, se observa una distribución tipo campana con un leve sesgo a la izquierda. Es lo que los estadísticos llamarían una distribución normal asimétrica.

Los datos atípicos pueden ocurrir por muchas razones. Quizás el estudiante se equivocó, ingresó 1 en vez de 10 ... o quizás es un estudiante muy perezoso y no le gusta estudiar. De cualquier forma, es una anomalía estadística que no representa a un estudiante típico. Miremos como se vería la distribución sin este.

En este caso omitimos el outlier (1) por razones de enseñanza. En la vida real es poco usual excluir valores en los extremos sin mayor justificación cuando nuestra muestra es muy reducida. Esto porque mientras más pequeña es la muestra, más probable es que sea una representación poco adecuada de toda la población, entendiendo esta como el conjunto de todos los estudiantes. Por ejemplo, si tomaramos las horas de estudio de otros 1000 estudiantes, nos daríamos cuenta que no es común estudiar poco.

Cuando tenemos más información la muestra se vuelve más confiable. Esto hace más fácil considerar estudiantes que caen por debajo o por ecnima de ciertos percentiles como outliers. Por ejemplo, el siguiente código usa la función __quantil__ de Pandas para excluir de la muestra observaciones por debajo del percentil 0.01, es decir por fuera de donde cae el 99 % de la muestra.

```python
q01 = df_estudiantes.HoraEstudio.quantile(0.01)

col = df_estudiantes[df_estudiantes.HoraEstudio>q01]['HoraEstudio']

show_distribution(col)
```

## Medidas de varianza:

Ahora sabemos como se comportan los valores medios de las horas de estudio y las calificaciones. Debemos proceder a examinar entonces otros aspectos de la distribución como: ¿Cuánta variabilidad hay en los datos?

Algunas medidas estadísticas de variabilidad son:

- __Rango__: diferencia entre el valor mínimo y el máximo.
- __Varianza__: promedio de la diferencia de cada datos de la media al cuadrado.
- __Desviación estándar__: raíz cuadrada de la varianza.

```python
for col_name in ['Calificacion','HoraEstudio']:
    col = df_estudiantes[col_name]
    rng = col.max() - col.min()
    var = col.var()
    std = col.std()
    print('\n{}:\n - Rango: {:.2f}\n - Varianza: {:.2f}\n - Desv.Est: {:.2f}'.format(col_name, rng, var, std))
```

De estos estadísticos el más útil es la _desviación estándar_ porque proporciona una medida de la varianza de los datos en la misma unidad de medida en que están los datos: unidades en términos de calificación y horas en horas de estudio. Nuevamente, parece existir un patrón discernible en el que los estudiantes que más estudian obtienen mejor calificación.

Esto lo podemos ver más claramente mediante una __línea de regresión__, al añadirla al gráfico muestra la tendencia general de los datos. Para hacer esto, usamos una técnica de regresión llamada __Mínimos cuadrados ordinarios__ que tiene la siguiente forma:

$ y = mx + b $

En esta ecuación __y__ y __x__ son las variables coordenadas, __m__ es la pendiente de la línea y __b__ es el intercepto en el eje y.

En el gráfico anterior __x__ corresponde a _HoraEstudio_ y __y__ a _Calificacion_, por lo que sólo necesitamos calcular el intercepto y la pendiente de la línea que pasa más cerca de todos esos puntos. Con esto, podemos formar una ecuación linear que calcule un nuevo valor de _y_ (Calificacion) para cada valor de _x_ (HoraEstudio).

Para evitar confusiones, este nuevo valor de _y_ lo llamamos _f(x)_ porque es el resultado de una función de ecuación linear basada en x.

La diferencia entre el _y_ original y el _f(x)_ es el error entre nuestra línea de regresión y la calificación real obtenida por el estudiante.

El objetivo es calcular la pendiente y el intercepto que minimize el promedio de los errores.

Afortunadamente, no tenemos que calcular esto nosotros mismos. El paquete __ScyPy__ incluye una clase __stats__ que proporciona un método __linregress__ que hace el trabajo por nosotros. Este arroja los coeficientes que necesitamos para la ecuación: la pendiente (_m_) y el intercepto (_b_) basado en el par de variables muestrales que queremos comparar.

```python
from scipy import stats


df_regresion = df_muestra[['Calificacion', 'HoraEstudio']].copy()


m, b, r, p, se = stats.linregress(df_regresion['HoraEstudio'], df_regresion['Calificacion'])
print('Pendiente: {:.4f}\nIntercepto-y: {:.4f}'.format(m,b))
print('Entonces ...\n f(x) = {:.4f}x + {:.4f}'.format(m,b))


df_regresion['fx'] = (m * df_regresion['HoraEstudio']) + b


df_regresion['error'] = df_regresion['fx'] - df_regresion['Calificacion']


df_regresion.plot.scatter(x='HoraEstudio', y='Calificacion')


plt.plot(df_regresion['HoraEstudio'],df_regresion['fx'], color='cyan')


plt.show()
```

Nótese que esta vez el gráfico tiene: la dispersión de las horas de estudio versus las calificaciones, y además, una línea que representa la ecuación de mínimos cuadrados ordinarios.

La pendiente y el intercepto calculados para la línea de regresión se muestran encima del gráfico.

La línea se basa entonces en los _f(x)_ calculados para cada __HoraEstudio__.

El siguiente código muestra:

- Las horas de estudio para cada estudiante.
- La calificación alcanzada por cada estudiante.
- El valor __f(x)__ calculado usando los coeficientes de la línea de regresión.
- El error entre el __f(x)__ calculado y la calificación real.

Algunos errores serán grandes (recordar que la correlación es 0.76).

```python
df_regresion[['HoraEstudio', 'Calificacion', 'fx', 'error']]
```

## Usando los coeficientes de regresión para realizar predicciones:

Ahora que tenemos los coeficientes de regresión para la relación entre las horas de estudio y la calificación, podemos usarlas en una función para estimar la calificación esperada para una cantidad de estudio determinada:

```python

def f(x):
    m = 0.3025
    b = 0.7056
    return m*x + b

tiempo_estudio = 7

# Get f(x) for study time
prediccion = f(tiempo_estudio)


calificacion_esperada = max(0,min(5,prediccion))


print ('Estudiar por {} horas a la semana lleva a una calificación esperada de {:.0f}'.format(tiempo_estudio, calificacion_esperada))
```

Entonces, aplicando estadística a la muestra determinamos la relación entre las horas de estudio y la calificación. Encapsulamos esta relación en una función general que podemos usarla para predecir la calificación de un estudiante de acuerdo con las horas semanales que dedique a estudiar.

Esta técnica es __la premisa de Machine Learning__, podemos tomar un conjunto de datos muestrales que incluya una o más _características_ (en este caso, el número de horas estudiada) y una _etiqueta_ conocida (en este caso la calificación) y usar los datos muestrales para obtener una función que calcule etiquetas predecidas para cada combinación dada de características.

## Modelos de regresión:

_Regresión_ es donde los modelos predicen números.

## ¿Qué es la regresión?

Es la relación entre variables de una base de datos que representan características (conocidas como _features_) del objeto de estudio, y la variable que tratamos de predecir (conocida como _label_).

Hagamos un ejemplo simple: Tenemos una empresa que alquila bicicletas en Medellín y queremos predecir el número esperado de alquileres en un día cualquiera. En este caso las características (_features_) pueden incluir variables como el día de la semana, el mes, etc. La etiqueta (_label_) es el número de bicicletas alquiladas en un día dado.

Para entrenar el modelo, comenzamos con una base de datos simple que contenga las características así como valores conocidos para las etiquetas. En este caso: información histórica que incluya fechas, temperatura y número de bicicletas arrendadas.

Después, dividimos la base de datos en dos _subsets_ (subconjuntos):

- __Una base de entrenamiento (_training dataset_)__ sobre el cual aplicaremos el algoritmo (función) que determina la función que contiene la relación entre las variables características y las etiquetas conocidas.
- __Una base de prueba (_test dataset_)__ que podemos usar para evaluar el modelo, usándola para crear predicciones para las etiquetas a partir del modelo y compararla después con la etiqueta real.

Este uso de bases de datos históricas, con etiquetas conocidas para entrenar modelos es lo que convierte a la regresión en una clase de __Machine Learning _supervisado___

### Machine learning supervisado:

Podemos pensarlo como una función en la que __y__ representa la etiqueta que queremos predecir y __x__ representa las características que el modelo usa para predecirla:

$$y = f(x)$$

En muchos casos, _x_ es un vector que consiste de múltiples características, por lo tanto la función toma esta forma:

$$y = f([x_1, x_2, x_3, ...])$$

Entonces, el objetivo de entrenar el modelo e encontrar la función que al realizar un cálculo a los valores _x_, llegue al producto _y_. Hacemos esto mediante un algoritmo de machine learning, cómo veremos. Hay muchos tipos de algoritmos de machine learning para el aprendizaje no supervisado, en términos generales los podemos dividir en dos tipos:

- __Algoritmos de regresión:__ algoritmos que predicen un valor _y_ que es numérico.
- __Algoritmos de clasificación:__ algoritmos que predicen la categoría o clase a la que pertenece una observación. El valor _y_ en este tió de modelos es un vector de probabilidades entre 0 y 1, uno para cada clase, indicando la probabilidad de pertenecer a cada clase.

## Entrenemos el modelo de regresión:

Lo primero que debemos hacer es seleccionar entonces aquellos posibles predictores del alquiler de bicicletas que identificamos en el análisis descriptivo.

Vamos a separar las características que usaremos para entrenar el modelo de las etiquetas que queremos predecir:

```python
X, y = bike_data[['season','mnth', 'holiday','weekday','workingday','weathersit','temp', 'atemp', 'hum', 'windspeed']].values, bike_data['rentals'].values

print('Features:',X[:10], '\nLabels:', y[:10], sep='\n')
```

Ahora tenemos dos matrices de Numpy llamadas:

- __X:__ características.
- __Y:__ etiquetas.

Podríamos entrenar el modelo con estos datos, pero es buena práctica en un modelo de aprendizaje supervisado dividir la base de datos en dos:

- Un subconjunto, tipicamente más largo, que nos permite entrenar el modelo.
- Un subconjunto, más pequeño, que usaremos para validar el modelo entrenado.

Esta buena práctica nos permite evaluar que tal se desempeña el modelo cuando se aplica en el subconjunto de validación (test), comparando la predicción con la etiqueta actual.

Es importante dividir la data de __forma aleatoria__. Esto permite que los dos subconjuntos sean estadísticamente comparables. Para esto validamos que la data de entrenamiento y la de validación tengan una distribución similar.

Esta división aleatoria la hacemos con la función __train_test_split__ de la librería __scikit-learn__. Esta es una de las librerías más usada para los procesos de Machine Learning en Python:


```python
from sklearn.model_selection import train_test_split

# Dividamos la base 70 % y 30 %
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)

print ('Training Set: %d rows\nTest Set: %d rows' % (X_train.shape[0], X_test.shape[0]))
```

Ahora tenemos cuatro bases de datos:

- __X_train:__ variables características que se usarán para entrenar el modelo.
- __y_train:__ variable etiqueta que se usara para entrenar el modelo.
- __X_test:__ variables características que se usarán para validar el modelo.
- __y_test:__ variable etiqueta que se usara para validar el modelo.

Ahora vamos a entrenar el modelo encontrando el algoritmo de regresión que mejor se ajuste a los datos de entrenamiento. Usaremos un algoritmo de _regresión lineal_, el punto inicial para las regresiones y que intenta encontrar una relación lineal entre los valores de _X_ y la etiqueta _y_. 

En Scikit-Learn los algoritmos de entrenamiento se encapsulan como _estimadores_, en este caso usaremos el estimador _LinearRegression_ para entrenar un modelo de regresión lineal:

```python
# Entrenar el modelo
from sklearn.linear_model import LinearRegression

# Encontrar un modelo de regresión lineal que se ajuste a los datos de entrenamiento:
model = LinearRegression().fit(X_train, y_train)
print (model)
```

### Evaluemos el modelo entrenado:

Ahora que hemos entrenado el modelo, podemos usarlo para predecir el número de alquileres de las características que retenimos en nuestro conjunto de datos de validación. Después, podemos comparar estas predicciones con los valores reales de la etiqueta para evaluar el rendimiento del modelo.

```python

predictions = model.predict(X_test)
np.set_printoptions(suppress=True)
print('Predicted labels: ', np.round(predictions)[:10])
print('Actual labels   : ' ,y_test[:10])
```

Comparar cada predicción con su valor real de base no es una forma muy eficiente de determinar la precisión de las predicciones del modelo. Veamos si podemos obtener una mejor indicación visualizando un diagrama de dispersión que compare las predicciones con las etiquetas reales. 

También superpondremos una línea de tendencia para obtener una idea general de la precisión con la que las etiquetas predichas se alinean con el laboratorio real.

```python


%matplotlib inline

plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Daily Bike Share Predictions')
# overlay the regression line
z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='magenta')
plt.show()
```

Existe una tendencia diagonal definida, y las intersecciones de los valores predichos y reales generalmente siguen la trayectoria de la línea de tendencia, pero existe una diferencia considerable entre la función ideal representada por la línea y los resultados. Esta varianza representa los residuos del modelo; en otras palabras, la diferencia entre la etiqueta predicha cuando el modelo aplica los coeficientes aprendidos durante el entrenamiento a los datos de validación y el valor real de la etiqueta de validación. Estos residuos, al evaluarse a partir de los datos de validación, indican el nivel de error esperado cuando el modelo se utiliza con datos nuevos cuya etiqueta se desconoce.

Los residuos se pueden cuantificar calculando varias métricas de evaluación de uso común. Nos centraremos en las tres siguientes:

- __Error cuadrático medio (MSE)__: La media de las diferencias al cuadrado entre los valores predichos y reales. Esto produce una métrica relativa en la que cuanto menor sea el valor, mejor se ajustará el modelo.
- __Error cuadrático medio (RMSE)__: La raíz cuadrada del ECM. Esto genera una métrica absoluta en la misma unidad que la etiqueta (en este caso, el número de alquileres). Cuanto menor sea el valor, mejor será el modelo (en resumen, representa el número promedio de alquileres con un margen de error en las predicciones).
- __Coeficiente de Determinación (generalmente conocido como R² o R²):__ Una métrica relativa en la que cuanto mayor sea el valor, mejor se ajustará el modelo. En esencia, esta métrica representa qué porcentaje de la varianza entre los valores predichos y los reales de la etiqueta puede explicar el modelo.


Calculemos estas métricas para nuestro modelo basados en las predicciones que genero para la base de validación:


```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)

rmse = np.sqrt(mse)
print("RMSE:", rmse)

r2 = r2_score(y_test, predictions)
print("R2:", r2)
```

El modelo tiene cierta capacidad predictiva ... pero podría ser mejor.

Después, intentaremos otros algoritmos de regresión para mejorar su desempeño.

### Algoritmos de conjunto:

Los algoritmos de conjunto funcionan combinando múltiples estimadores base para producir un modelo óptimo, ya sea aplicando una función de agregación a un conjunto de modelos base (lo que a veces se denomina "bagging") o construyendo una secuencia de modelos que se complementan para mejorar el rendimiento predictivo (lo que se denomina "boosting").

Por ejemplo, probemos un modelo de Bosque Aleatorio, que aplica una función de promedio a múltiples modelos de Árbol de Decisión para obtener un mejor modelo general.

```python
from sklearn.ensemble import RandomForestRegressor

# Train the model
model = RandomForestRegressor().fit(X_train, y_train)
print (model, "\n")

# Evaluate the model using the test data
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_test, predictions)
print("R2:", r2)

# Plot predicted vs actual
plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Daily Bike Share Predictions')
# overlay the regression line
z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='magenta')
plt.show()
```

```python
# Train the model
from sklearn.ensemble import GradientBoostingRegressor

# Fit a lasso model on the training set
model = GradientBoostingRegressor().fit(X_train, y_train)
print (model, "\n")

# Evaluate the model using the test data
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_test, predictions)
print("R2:", r2)

# Plot predicted vs actual
plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Daily Bike Share Predictions')
# overlay the regression line
z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='magenta')
plt.show()
```

### Preprocesando los datos:

Esto permite que algoritmo ajuste los modelos de forma más eficiente. Estas son algunas de las técnicas más comunes:

- Escalar características numéricas.
- Codificar variables categóricas.
- "One-hot encoding".

Para aplicar estas transformaciones de preprocesamiento al alquiler de bicicletas, utilizaremos una función de Scikit-Learn llamada *pipelines*. Estas canalizaciones nos permiten definir un conjunto de pasos de preprocesamiento que culminan en un algoritmo. Posteriormente, se puede ajustar toda la canalización a los datos, de modo que el modelo encapsule todos los pasos de preprocesamiento, así como el algoritmo de regresión. 

Esto es útil, ya que, al usar el modelo para predecir valores a partir de nuevos datos, necesitamos aplicar las mismas transformaciones (basadas en las mismas distribuciones estadísticas y codificaciones de categorías utilizadas con los datos de entrenamiento).

```python
# Train the model
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LinearRegression
import numpy as np

# Define preprocessing for numeric columns (scale them)
numeric_features = [6,7,8,9]
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())])

# Define preprocessing for categorical features (encode them)
categorical_features = [0,1,2,3,4,5]
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

# Create preprocessing and training pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('regressor', GradientBoostingRegressor())])


# fit the pipeline to train a linear regression model on the training set
model = pipeline.fit(X_train, (y_train))
print (model)
```

```python
# Get predictions
predictions = model.predict(X_test)

# Display metrics
mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_test, predictions)
print("R2:", r2)

# Plot predicted vs actual
plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Daily Bike Share Predictions')
z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='magenta')
plt.show()
```

Ensayemos con otro algoritmo:

```python
# Use a different estimator in the pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('regressor', RandomForestRegressor())])


# fit the pipeline to train a linear regression model on the training set
model = pipeline.fit(X_train, (y_train))
print (model, "\n")

# Get predictions
predictions = model.predict(X_test)

# Display metrics
mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_test, predictions)
print("R2:", r2)

# Plot predicted vs actual
plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Daily Bike Share Predictions - Preprocessed')
z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='magenta')
plt.show()
```
