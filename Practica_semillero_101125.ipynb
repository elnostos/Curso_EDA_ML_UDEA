## Seleccionar la(s) variable(s) que considere más importante(s) en la base de datos, plantear al menos un objetivo y realizar un análisis descriptivo de dicha(s) variable(s).

Pregunta: **¿Cómo se comportan los homicidios en Medellín según la modalidad (arma)?**

- **Variable principal:** modalidad.
- **Variables secundarias:** edad, sexo, comuna, jornada (no existe).

**¿Por qué podrían ser covariables adecuadas?**

**¿Cómo creo una variable a partir de otras?:** jornada del hecho

```python
# Crear variable jornada:

# hora_hecho debe ser numérica
df["hora_hecho"] = pd.to_numeric(df["hora_hecho"], errors="coerce")

# Crear la variable jornada
df["jornada"] = pd.cut(
    df["hora_hecho"],
    bins=[-1, 5, 11, 18, 23],
    labels=["Madrugada", "Mañana", "Tarde", "Noche"],
    include_lowest=True
)
```

```python
df[["hora_hecho", "jornada"]].head(10)
```

**Miremos la distribución de la modalidad entre categorías**:

```python
df["modalidad"].value_counts()
```

```python
round(df["modalidad"].value_counts(normalize=True) * 100,2)
```

¿Que tal si creamos una variable llamada `modalidad1` que tenga las primeras 5 categorías de modalidad y el resto agrupadas?

```python
top5 = df["modalidad"].value_counts().nlargest(5).index

df["modalidad1"] = df["modalidad"].where(df["modalidad"].isin(top5), "Resto")

df["modalidad1"].value_counts()
```

¿**Gráficamente**, como podemos mirar la distribución de `modalidad1` entre sus categorías?

```python
import matplotlib.pyplot as plt

conteo = df["modalidad1"].value_counts()

plt.figure(figsize=(8, 6))
conteo.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Distribución de la modalidad del hecho")
plt.xlabel("Modalidad")
plt.ylabel("Cantidad de casos")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

**Una categoría tiene un nombre muy largo, podemos hacer dos cosas:** veamos como podemos manejar esto

```python
df["modalidad1"] = df["modalidad1"].replace(
    {"Ahorcamiento o estrangulamiento": "Ahorcamiento"}
)
```

```python
df["modalidad1"] = df["modalidad1"].replace({
    "Asfixia mecánica": "Asfixia\nmecánica",
    "Arma de fuego": "Arma de\nfuego"
})
```

### Ahora analicemos modalidad con la edad:

¿Qué gráfico es más adecuado para comparar variables categóricas y continuas?

```python
import pandas as pd
import matplotlib.pyplot as plt

homicidios = (
    df.groupby(["edad", "sexo"])
    .size()
    .reset_index(name="cantidad")
    .sort_values("edad")
)

plt.figure(figsize=(9, 6))

for sexo, datos in homicidios.groupby("sexo"):
    plt.plot(
        datos["edad"],
        datos["cantidad"],
        marker="o",
        linestyle="-",
        label=sexo
    )

plt.title("Distribución de homicidios por edad y sexo", fontsize=14, weight="bold")
plt.xlabel("Edad")
plt.ylabel("Número de homicidios")
plt.legend(title="Sexo o género")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
