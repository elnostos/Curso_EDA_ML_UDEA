## 1.1 Descargar e instalar Anaconda

- _Anaconda_ es un programa que instala _Python_, _Jupyter Notebook_ y muchas herramientas para ciencia de datos en un solo paquete.

- Nos permite trabajar con entornos virtuales para que cada proyecto tenga sus propias librerías y versiones de Python.

- Incluye todas las librerías básicas para el curso: numpy, pandas, matplotlib, etc.


__Pasos:__

1. Ve a https://www.anaconda.com/download

2. Selecciona el instalador de tu sistema operativo (Windows / macOS / Linux).

3. Descarga la versión para Python 3.x (actualmente 3.12 o 3.11, no la 2.7).

4. Guarda el archivo en tu computador.

5. Ejecuta el instalador:

- En Windows:

        Marca la opción "Add Anaconda to my PATH environment variable" solo si quieres usar Python desde cualquier terminal (opcional, pero útil).

Acepta "Register Anaconda as my default Python".

- En macOS/Linux:

        Abre el archivo .pkg o usa el terminal para instalar, sigue los pasos del asistente y finaliza la instalación.

6. Verifica la instalación:

Abrir:

- Windows: "Anaconda Prompt"

- macOS/Linux: Terminal

- Escribir: _conda --version_ y _python --version_

- Si todo va bien deberíamos ver algo como: _conda 24.5.0_, _Python 3.12.2_

7. Instalar Jupyter Notebook:

Con el entorno activo escribir: _conda install jupyter_

8. Abrimos Jupyter:

Con el entorn activo escribir: _jupyter notebook_

La analítica de datos es el proceso sistemático de recolectar, organizar, describir, analizar e interpretar información para comprender fenómenos, responder preguntas y apoyar la toma de decisiones basadas en evidencia.

Incluye:

1. Etapa descriptiva (estadística descriptiva):

- Resumir y presentar los datos de manera clara mediante medidas numéricas (media, mediana, proporciones) y representaciones gráficas (tablas, diagramas, mapas).

- El objetivo aquí es entender qué dicen los datos tal como son.

2. Etapa inferencial o predictiva (estadística inferencial y métodos de modelado):

- Usar muestras para sacar conclusiones sobre una población más amplia.

- Construir modelos que permitan explicar relaciones y, en algunos casos, predecir resultados futuros.

3. Comunicación de hallazgos:

- Convertir los resultados en un relato claro, comprensible y útil para el público objetivo (en nuestro caso, decisores políticos, ciudadanía o comunidad académica).

__Machine Learning__ (aprendizaje automático) es, en pocas palabras, un conjunto de métodos algorítmicos y estadísticos que permiten a una computadora aprender patrones a partir de datos y luego hacer predicciones o tomar decisiones sin estar programada explícitamente para cada caso. En lugar de decirle paso a paso qué hacer, le damos ejemplos y la máquina aprende las reglas por sí sola.

_Arthur Samuel_, el pionero de la inteligencia artificial y el Machine Learning definió esta última, en 1959, cómo:

    "El campo de estudio que le da a las computadoras la habilidad de aprender sin ser explícitamente programadas".

En los y las autoras de la bibliografía se asumen tres perspectivas con respecto al Machine Learning:


__Definición técnica (enfoque de ingeniería)__
Según Aurélien Géron, es el campo de la inteligencia artificial que entrena sistemas para mejorar automáticamente con la experiencia (datos), abarcando técnicas como regresión, clasificación, clustering, redes neuronales, etc. Su objetivo es optimizar una función de rendimiento y generalizar bien a nuevos datos.

__Definición cultural y aplicada (enfoque en ciencias sociales)__
Para Grimmer, Roberts y Stewart, es una “cultura” con herramientas y valores propios, centrada muchas veces en la predicción y en trabajar con grandes volúmenes y variedad de datos, aplicable no solo para predecir, sino también para descubrir conceptos, medir fenómenos y estimar efectos causales.

__Visión crítica y social__
Cathy O’Neil recuerda que los modelos de Machine Learning no son neutrales: pueden reproducir sesgos y generar efectos negativos a gran escala si son opacos, no se auditan y se aplican sin cuidado. Los llama “armas de destrucción matemática” cuando amplifican desigualdades y no rinden cuentas.

En este curso introduciremos varios tipos de Machine Learning que pueden ser clasificados en las siguientes categorías:

### Aprendizaje supervisado (Supervised learning):

- El modelo aprende a partir de un conjunto de datos etiquetados: cada ejemplo de entrenamiento incluye las _características_ (inputs) y la _etiqueta_ o valor objetivo (output).
- El objetivo es predecir la etiqueta de nuevos datos no vistos.
- Ejemplos: __regresión lineal__, __regresión logística__, __árboles de decisión__, __k-NN__, SVM, redes neuronales supervisadas.
- Casos de uso: predicción de precios de vivienda, clasificación de correos como spam/no spam.

### Aprendizaje NO supervisado (Unsupervised learning):

- El modelo recibe datos sin etiquetas y debe encontrar patrones, estructuras o relaciones ocultas.
- Ejemplos: __clustering (K-Means, DBSCAN)__, __detección de anomalías__, _reducción de dimensionalidad (PCA, t-SNE)_, modelos de mezcla (Gaussian Mixtures).
- Casos de uso: segmentación de clientes, compresión de datos, detección de anomalías.

### Aprendizaje por refuerzo (Reinforcement Learning):

- Un agente aprende a tomar decisiones en un entorno para maximizar una recompensa acumulada.
- Se basa en la interacción acción → estado → recompensa → retroalimentación.
- Ejemplos: Q-Learning, Deep Q-Networks (DQN), métodos de política (Policy Gradient).
- Casos de uso: robótica, juegos (AlphaGo), control de sistemas.
