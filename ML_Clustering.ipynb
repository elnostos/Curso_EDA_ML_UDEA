## Machine Learning no supervisado

### ¿Qué es Machine Learning no supervisado?

El **Machine Learning no supervisado** es un tipo de aprendizaje automático donde el algoritmo encuentra patrones en los datos **sin etiquetas predefinidas**. A diferencia del aprendizaje supervisado (donde tienes ejemplos etiquetados como "democracia" o "autocracia"), aquí el algoritmo descubre la estructura por sí mismo.

**Analogía simple**: 
- **Supervisado**: Le muestras fotos etiquetadas "perro" y "gato" para que aprenda a distinguirlos
- **No supervisado**: Le das un montón de fotos sin etiquetas y encuentra que hay dos tipos de animales similares

### Tipos de Machine Learning no supervisado

Existen varios tipos de técnicas no supervisadas:

#### 1. **Clustering (agrupamiento)**
- **Qué hace**: Agrupa datos similares en grupos o "clusters"
- **Ejemplo en ciencia política**: Agrupar ciudades colombianas según sus políticas públicas
- **Cuándo usar**: Cuando quieres descubrir tipologías o clasificaciones no predefinidas
- **Métodos comunes**: K-Means, clustering jerárquico

#### 2. **Reducción de dimensionalidad**
- **Qué hace**: Reduce el número de variables manteniendo la información importante
- **Ejemplo en ciencia política**: Simplificar múltiples indicadores políticos en dimensiones principales
- **Cuándo usar**: Cuando tienes muchas variables y quieres simplificar
- **Método más comun**: PCA (Análisis de Componentes Principales)

#### 3. **Detección de anomalías (outliers)**
- **Qué hace**: Identifica casos que son muy diferentes del resto
- **Ejemplo en ciencia política**: Encontrar ciudades con políticas muy atípicas
- **Cuándo usar**: Cuando quieres identificar casos excepcionales
- **Métodos comunes**: Isolation Forest

#### 4. **Asociación**
- **Qué hace**: Encuentra relaciones entre variables o eventos
- **Ejemplo en ciencia política**: ¿Qué políticas públicas tienden a aparecer juntas?
- **Cuándo usar**: Cuando quieres entender relaciones entre variables
- **Métodos comunes**: Reglas de asociación, análisis de mercado

## Conceptos fundamentales (sin matemáticas complicadas)

### ¿Cómo funciona el clustering?

El clustering funciona con un principio muy simple: **"Las cosas similares están cerca, las diferentes están lejos"**.

#### Concepto 1: Distancia
- En lugar de medir distancia en kilómetros, medimos **similitud entre datos**
- Si dos ciudades (o casos) tienen políticas muy similares, están "cerca"
- Si tienen políticas muy diferentes, están "lejos"
- **Ejemplo**: Medellín y Bogotá podrían estar "cerca" si tienen patrones de gasto público similares, pero "lejos" de una ciudad pequeña con políticas muy diferentes

#### Concepto 2: Centroides (Centros de grupos)
- Cada grupo tiene un "centro" que representa las características promedio del grupo
- Los casos se agrupan alrededor de estos centros
- **Pensémoslo así**: Si tienes un grupo de ciudades con "alto gasto en educación", el centroide sería una ciudad "promedio" con esas características

#### Concepto 3: Número de grupos (k)
- Debemos decidir cuántos grupos queremos encontrar
- Demasiados grupos = cada caso en su propio grupo (no útil, no hay generalización)
- Muy pocos grupos = todos mezclados (no útil, no hay diferenciación)
- El objetivo es encontrar el número "justo" que capture diferencias reales pero sea interpretable

---

## ¿Cómo funciona K-Means? (explicación detallada)

K-Means es el algoritmo más común:

### El proceso en 4 pasos:

#### **Paso 1: Inicialización**
- El algoritmo elige aleatoriamente **k puntos** como "centros iniciales" de los grupos
- Imagina que lanzas k dardos al azar en un mapa de tus datos
- **Ejemplo**: Si queremos 3 grupos de ciudades colombianas, elegimos 3 ciudades al azar como "centros iniciales"

#### **Paso 2: Asignación**
- Cada punto (ciudad, caso) se asigna al centro más cercano
- **Ejemplo**: Medellín se asigna al centro más cercano. Si el centro representa "ciudades con alto gasto en educación", Medellín se une a ese grupo

#### **Paso 3: Actualización**
- Se calculan nuevos centros basados en los puntos asignados a cada grupo
- El nuevo centro es el "promedio" de todos los puntos en ese grupo
- **Ejemplo**: Si 5 ciudades están en un grupo, el nuevo centro es el punto promedio de esas 5 ciudades

#### **Paso 4: Repetición**
- Se repiten los pasos 2 y 3 hasta que los centros ya no cambien (o cambien muy poco)
- Esto significa que el algoritmo "convergió" y encontró los grupos óptimos

### Visualización del proceso:

```
Iteración 1: Centros aleatorios → Asignación → Nuevos centros
Iteración 2: Nuevos centros → Re-asignación → Centros actualizados
Iteración 3: Centros actualizados → Re-asignación → Centros finales
...
Iteración N: Los centros ya no cambian → ¡Listo!
```

### ¿Por qué se llama "K-Means"?
- **K**: El número de grupos que especificamos
- **Means**: "Medias" en inglés - el algoritmo usa promedios (medias) para calcular los centros

## Ejemplo con 2 variables:

Vamos a empezar con el ejemplo más simple posible: **solo 2 variables**. Esto nos permite ver los grupos en un gráfico de 2 dimensiones (como un mapa).

### Escenario de investigación

Imaginemos que queremos analizar ciudades colombianas según:
- **Variable 1**: Gasto público en educación (% del presupuesto municipal)
- **Variable 2**: Gasto público en salud (% del presupuesto municipal)

¿Podemos encontrar grupos de ciudades con políticas de gasto similares?


````python
# Crear datos de ejemplo (simulados para ciudades colombianas)
# En la práctica, estos datos vendrían de bases de datos reales como:
# - DANE (Departamento Administrativo Nacional de Estadística)
# - Datos de alcaldías y gobernaciones

# Simulamos datos para ciudades colombianas principales
np.random.seed(42)  # The Answer to the Ultimate Question of Life, the Universe, and Everything

# Grupo 1: Ciudades con alto gasto en educación y salud
# Típicamente ciudades grandes con mayor capacidad fiscal
ciudades_alto_gasto = {
    'ciudad': ['Bogotá', 'Medellín', 'Cali', 'Barranquilla'],
    'gasto_educacion': np.random.normal(4.8, 0.3, 4),  # Alto gasto en educación
    'gasto_salud': np.random.normal(5.8, 0.3, 4)      # Alto gasto en salud
}

# Grupo 2: Ciudades con gasto medio
# Ciudades intermedias con recursos moderados
ciudades_gasto_medio = {
    'ciudad': ['Bucaramanga', 'Pereira', 'Manizales', 'Armenia', 'Ibagué',
               'Santa Marta', 'Valledupar', 'Montería', 'Villavicencio', 'Pasto'],
    'gasto_educacion': np.random.normal(3.8, 0.4, 10),  # Gasto medio en educación
    'gasto_salud': np.random.normal(4.8, 0.4, 10)        # Gasto medio en salud
}

# Grupo 3: Ciudades con gasto bajo
# Ciudades más pequeñas o con recursos limitados
ciudades_bajo_gasto = {
    'ciudad': ['Cartagena', 'Sincelejo', 'Tunja', 'Popayán', 'Neiva',
               'Riohacha', 'Quibdó', 'Florencia', 'Mocoa', 'Leticia',
               'San Andrés', 'Yopal', 'Arauca', 'Bello', 'Mitú'],
    'gasto_educacion': np.random.normal(3.2, 0.5, 15),  # Gasto bajo en educación
    'gasto_salud': np.random.normal(4.2, 0.5, 15)        # Gasto bajo en salud
}

# Combinar todos los datos
datos_alto = pd.DataFrame(ciudades_alto_gasto)
datos_medio = pd.DataFrame(ciudades_gasto_medio)
datos_bajo = pd.DataFrame(ciudades_bajo_gasto)

# Crear un solo DataFrame
datos_ciudades = pd.concat([datos_alto, datos_medio, datos_bajo], ignore_index=True)


print(f"Total de ciudades: {len(datos_ciudades)}")
print(f"  - Ciudades con alto gasto: {len(datos_alto)}")
print(f"  - Ciudades con gasto medio: {len(datos_medio)}")
print(f"  - Ciudades con gasto bajo: {len(datos_bajo)}")
print("\nPrimeras filas de los datos:")
print(datos_ciudades.head(10))
print("\nTodas las ciudades incluidas:")
print(datos_ciudades[['ciudad', 'gasto_educacion', 'gasto_salud']].sort_values('ciudad'))
````

### Visualización: Ver los datos antes del clustering

Antes de hacer clustering, es buena práctica **visualizar los datos** para tener una idea de qué esperar.


````python
# Crear un gráfico de dispersión (scatter plot)
plt.figure(figsize=(10, 6))
plt.scatter(datos_ciudades['gasto_educacion'], 
            datos_ciudades['gasto_salud'],
            s=100, alpha=0.6, edgecolors='black', linewidth=1)

# Agregar etiquetas a algunas ciudades para referencia
for i, ciudad in enumerate(datos_ciudades['ciudad']):
    if ciudad in ['Bogotá', 'Medellín', 'Cali', 'Barranquilla', 'Bucaramanga', 
                  'Pereira', 'Cartagena', 'Manizales']:
        plt.annotate(ciudad, 
                    (datos_ciudades['gasto_educacion'].iloc[i], 
                     datos_ciudades['gasto_salud'].iloc[i]),
                    fontsize=9, alpha=0.7)

plt.xlabel('Gasto en educación (% del presupuesto)', fontsize=12, fontweight='bold')
plt.ylabel('Gasto en salud (% del presupuesto)', fontsize=12, fontweight='bold')
plt.title('Distribución de ciudades por gasto público\n(ANTES del clustering)', 
          fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("Observación: ¿Ves cómo las ciudades parecen agruparse en ciertas zonas?")
print("El clustering nos ayudará a identificar estos grupos de forma automática.")
````

### Aplicar K-Means Clustering

Ahora vamos a aplicar el algoritmo K-Means. Necesitamos decidir cuántos grupos queremos. Empecemos con **3 grupos** (parece razonable basándonos en el gráfico).


````python
# Paso 1: Preparar los datos para el clustering
# Solo necesitamos las columnas numéricas (no el nombre de la ciudad)
X = datos_ciudades[['gasto_educacion', 'gasto_salud']].values

# Paso 2: Aplicar K-Means con 3 grupos
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
grupos = kmeans.fit_predict(X)  # Esto asigna cada ciudad a un grupo (0, 1, o 2)

# Paso 3: Agregar la información del grupo a nuestros datos
datos_ciudades['grupo'] = grupos

print("¡Clustering completado!")
print(f"\nDistribución de ciudades por grupo:")
print(datos_ciudades['grupo'].value_counts().sort_index())
print("\nPrimeras ciudades con su grupo asignado:")
print(datos_ciudades[['ciudad', 'gasto_educacion', 'gasto_salud', 'grupo']].head(10))
````

### Visualizar los Resultados del Clustering

Ahora vamos a ver cómo quedaron agrupados los países. Cada color representa un grupo diferente.


````python
# Crear gráfico con los grupos identificados
plt.figure(figsize=(12, 7))

# Colores para cada grupo
colores = ['#FF6B6B', '#7d2181', '#95E1D3']
nombres_grupos = ['Grupo 1: Gasto bajo', 'Grupo 2: Gasto alto', 'Grupo 3: Gasto medio']

# Graficar cada grupo con un color diferente
for i in range(3):
    grupo_data = datos_ciudades[datos_ciudades['grupo'] == i]
    plt.scatter(grupo_data['gasto_educacion'], 
                grupo_data['gasto_salud'],
                c=colores[i], 
                label=nombres_grupos[i],
                s=100, alpha=0.7, edgecolors='black', linewidth=1)

# Marcar los centros de cada grupo (centroides)
centroides = kmeans.cluster_centers_
plt.scatter(centroides[:, 0], centroides[:, 1],
            c='yellow', marker='X', s=300, 
            edgecolors='black', linewidth=2,
            label='Centros de los grupos', zorder=10)

# Agregar etiquetas a algunas ciudades
for i, ciudad in enumerate(datos_ciudades['ciudad']):
    if ciudad in ['Bogotá', 'Medellín', 'Cali', 'Barranquilla', 'Bucaramanga', 
                  'Pereira', 'Cartagena', 'Manizales', 'Armenia', 'Ibagué']:
        plt.annotate(ciudad, 
                    (datos_ciudades['gasto_educacion'].iloc[i], 
                     datos_ciudades['gasto_salud'].iloc[i]),
                    fontsize=8, alpha=0.8)

plt.xlabel('Gasto en educación (% del presupuesto)', fontsize=12, fontweight='bold')
plt.ylabel('Gasto en salud (% del presupuesto)', fontsize=12, fontweight='bold')
plt.title('Clustering de ciudades por gasto público\n(K-Means con 3 grupos)', 
          fontsize=14, fontweight='bold')
plt.legend(loc='best', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nInterpretación:")
print("- Los puntos amarillos con X son los 'centros' de cada grupo")
print("- Cada ciudad está asignada al grupo cuyo centro está más cerca")
print("- Las ciudades del mismo color tienen políticas de gasto similares")
````

### Analizar las Características de Cada Grupo

Ahora vamos a entender qué caracteriza a cada grupo. Esto es crucial para interpretar los resultados.


````python
# Analizar las características promedio de cada grupo
print("=" * 60)
print("CARACTERÍSTICAS DE CADA GRUPO")
print("=" * 60)

for i in range(3):
    grupo_data = datos_ciudades[datos_ciudades['grupo'] == i]
    print(f"\n{nombres_grupos[i]}:")
    print(f"  - Número de ciudades: {len(grupo_data)}")
    print(f"  - Gasto promedio en educación: {grupo_data['gasto_educacion'].mean():.2f}%")
    print(f"  - Gasto promedio en salud: {grupo_data['gasto_salud'].mean():.2f}%")
    print(f"  - Ciudades en este grupo:")
    for ciudad in grupo_data['ciudad'].values:
        print(f"    • {ciudad}")

print("\n" + "=" * 60)
print("INTERPRETACIÓN:")
print("=" * 60)
print("""
El algoritmo identificó automáticamente tres grupos de ciudades:

1. Grupo 1 (Gasto bajo): Ciudades con políticas de gasto más limitadas
   - Principalmente ciudades más pequeñas o con recursos limitados
   
2. Grupo 2 (Gasto alto): Ciudades con políticas de gasto social más generosas
   - Típicamente ciudades grandes con mayor capacidad fiscal
   
3. Grupo 3 (Gasto medio): Ciudades con políticas de gasto moderadas
   - Principalmente ciudades intermedias con recursos moderados

Esto permite comparar y categorizar ciudades sin necesidad de definir
a priori qué es "alto" o "bajo" gasto. El algoritmo lo descubrió por sí solo.
""")
````

## ¿Cómo elegir el número correcto de grupos?

Una pregunta importante: **¿Cómo sabemos que 3 grupos es el número correcto?**

En la práctica, no siempre es obvio. Vamos a explorar diferentes números de grupos y ver cuál funciona mejor.


## Entendiendo el método del codo (Elbow Method)

El **Método del Codo** es una técnica visual para elegir el número óptimo de grupos. Te explicamos cómo funciona:

### ¿Qué es la Inercia?

**Inercia** (también llamada "within-cluster sum of squares" o WCSS) mide qué tan "compactos" son los grupos.

- **Baja inercia** = Los puntos están muy cerca de su centro (grupos compactos)
- **Alta inercia** = Los puntos están lejos de su centro (grupos dispersos)

**Fórmula conceptual**: Suma de las distancias al cuadrado de cada punto a su centroide

### ¿Cómo Funciona el Método del Codo?

1. **Probamos diferentes valores de k** (2, 3, 4, 5, 6...)
2. **Calculamos la inercia** para cada k
3. **Graficamos**: k en el eje X, inercia en el eje Y
4. **Buscamos el "codo"**: El punto donde agregar más grupos ya no reduce mucho la inercia


**Antes del codo**: Agregar grupos reduce mucho la inercia (mejora significativa)
**Después del codo**: Agregar grupos reduce poco la inercia (mejora marginal)

### Ejemplo Práctico:

Si tenemos:
- **k=2**: Inercia = 100 (grupos muy dispersos)
- **k=3**: Inercia = 30 (gran mejora) ← Probablemente el codo
- **k=4**: Inercia = 25 (pequeña mejora)
- **k=5**: Inercia = 23 (mejora muy pequeña)

**Conclusión**: k=3 es probablemente óptimo porque después de eso, agregar más grupos no ayuda mucho.

### Ventajas del método del codo:

- **Visual e intuitivo**: Fácil de entender
- **Rápido**: No requiere cálculos complejos
- **Útil como punto de partida**: Te da una idea inicial

### Limitaciones:

- **Subjetivo**: A veces el "codo" no es claro
- **Puede no funcionar**: Si los datos no tienen grupos naturales, no habrá codo claro
- **No considera separación**: Solo mide compactitud, no qué tan separados están los grupos

### Combinar con Otras Métricas:

El método del codo es mejor cuando lo usas **junto con**:
- **Silhouette Score**: Mide separación entre grupos
- **Conocimiento teórico**: ¿Qué número de grupos tiene sentido para la pregunta?
- **Validación práctica**: ¿Los grupos son interpretables?

**En la práctica**: Usa el método del codo como guía inicial, pero valida con Silhouette Score y conocimiento teórico.


````python
# Probar diferentes números de grupos (de 2 a 6)
rango_k = range(2, 7)
inertias = []  # Mide qué tan "compactos" son los grupos
silhouettes = []  # Mide qué tan bien separados están los grupos

for k in rango_k:
    kmeans_temp = KMeans(n_clusters=k, random_state=42, n_init=10)
    grupos_temp = kmeans_temp.fit_predict(X)
    
    # Inercia: suma de distancias al cuadrado desde cada punto a su centro
    # Menor inercia = grupos más compactos
    inertias.append(kmeans_temp.inertia_)
    
    # Silhouette: mide qué tan bien separados están los grupos
    # Valores entre -1 y 1. Más cerca de 1 = mejor separación
    silhouette = silhouette_score(X, grupos_temp)
    silhouettes.append(silhouette)
    
    print(f"k={k}: Inercia={kmeans_temp.inertia_:.2f}, Silhouette={silhouette:.3f}")

# Visualizar los resultados
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico 1: Método del Codo (Elbow Method)
ax1.plot(rango_k, inertias, 'bo-', linewidth=2, markersize=8)
ax1.set_xlabel('Número de Grupos (k)', fontsize=11, fontweight='bold')
ax1.set_ylabel('Inercia', fontsize=11, fontweight='bold')
ax1.set_title('Método del Codo\n(Buscamos el "codo" en la curva)', 
              fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.set_xticks(rango_k)

# Gráfico 2: Silhouette Score
ax2.plot(rango_k, silhouettes, 'ro-', linewidth=2, markersize=8)
ax2.set_xlabel('Número de Grupos (k)', fontsize=11, fontweight='bold')
ax2.set_ylabel('Silhouette Score', fontsize=11, fontweight='bold')
ax2.set_title('Silhouette Score\n(Mayor = mejor separación)', 
              fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.set_xticks(rango_k)
ax2.axhline(y=0, color='gray', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

print("\nExplicación:")
print("- El 'Método del Codo' busca un punto donde agregar más grupos no mejora mucho")
print("- El 'Silhouette Score' mide qué tan bien separados están los grupos")
print("- En este caso, k=3 parece ser una buena elección")
````

## Entendiendo el Silhouette Score:

El **Silhouette Score** es una de las métricas más importantes para evaluar la calidad de nuestro clustering. Así es cómo funciona:

### ¿Qué mide el Silhouette Score?

El Silhouette Score mide **qué tan bien separados están los grupos** y **qué tan coherentes son internamente**. 

**Rango de valores**: De -1 a +1
- **+1**: Grupos perfectamente separados y muy coherentes (ideal)
- **0**: Los grupos están muy cerca o se superponen
- **-1**: Los puntos están asignados al grupo incorrecto

### ¿Cómo se calcula? (Concepto Simple)

Para cada punto (ciudad, caso), el algoritmo calcula:

1. **a(i)**: Distancia promedio a otros puntos **dentro del mismo grupo** (cohesión interna)
   - Si este valor es pequeño = el punto está bien agrupado con sus vecinos

2. **b(i)**: Distancia promedio al grupo **más cercano diferente** (separación)
   - Si este valor es grande = el punto está lejos de otros grupos

3. **Silhouette de ese punto**: `(b(i) - a(i)) / max(a(i), b(i))`
   - Si b(i) >> a(i) → el punto está bien asignado (score cercano a +1)
   - Si a(i) >> b(i) → el punto está mal asignado (score cercano a -1)

4. **Silhouette Score general**: Promedio de todos los puntos

### Interpretación Práctica:

| Score | Interpretación | ¿Qué hacer? |
|-------|----------------|-------------|
| **0.7 - 1.0** | Grupos muy bien definidos | Excelente, puedes confiar en los resultados |
| **0.5 - 0.7** | Grupos razonablemente definidos | Bueno, resultados son útiles |
| **0.25 - 0.5** | Grupos algo difusos | Considera revisar el número de grupos o las variables |
| **< 0.25** | Grupos muy mal definidos | Los grupos no son claros, reconsidera tu análisis |

### Ejemplo:

Imagina que Medellín tiene:
- **a(Medellín)**: Distancia promedio a otras ciudades en su grupo = 0.3
- **b(Medellín)**: Distancia al grupo más cercano = 0.8
- **Silhouette**: (0.8 - 0.3) / 0.8 = 0.625 (Bueno)

Esto significa que Medellín está:
- Cerca de otras ciudades en su grupo (cohesión)
- Lejos de ciudades en otros grupos (separación)

### Análisis de Perfiles de Cada Grupo

Vamos a crear un perfil promedio de cada grupo para entender mejor qué los caracteriza.


#### Crear y visualizar el dendrograma

Ahora vamos a crear un dendrograma para visualizar la estructura jerárquica de nuestros datos. Esto nos permitirá ver todos los niveles de agrupación posibles.

**Nota**: El dendrograma es especialmente útil cuando trabajas con un número moderado de casos (menos de 100-200), ya que con muchos casos puede volverse difícil de leer.


````python
# Para crear el dendrograma, necesitamos importar linkage y dendrogram de scipy.cluster.hierarchy
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import numpy as np

# Usaremos los datos normalizados del ejemplo complejo
# 'ward' es un método común de linkage que minimiza la varianza dentro de los grupos

# Calcular la matriz de linkage
Z = linkage(X_normalizado, method='ward')

# Calcular dinámicamente la altura de corte para obtener 3 grupos
# Usamos un método iterativo para encontrar la altura que produce exactamente 3 grupos
n_puntos = len(datos_complejos)
k_grupos = 10

# Obtener todas las alturas de fusión
alturas = Z[:, 2]

# Buscar la altura que produce k grupos
# Empezamos desde la altura mínima y vamos aumentando
altura_min = alturas.min()
altura_max = alturas.max()

# Usar búsqueda binaria o simplemente probar alturas
# Método simple: probar alturas en el rango y encontrar la que da k grupos
altura_corte = altura_max
for altura_prueba in np.linspace(altura_min, altura_max, 100):
    grupos_prueba = fcluster(Z, altura_prueba, criterion='distance')
    n_grupos_unicos = len(set(grupos_prueba))
    if n_grupos_unicos == k_grupos:
        altura_corte = altura_prueba
        break
    elif n_grupos_unicos < k_grupos:
        # Si tenemos menos grupos de los deseados, usar la altura anterior
        break

# Si no encontramos exactamente k grupos, usar la altura que produce el número más cercano
if len(set(fcluster(Z, altura_corte, criterion='distance'))) != k_grupos:
    # Usar la altura de la fusión que deja k grupos
    # Para k grupos, necesitamos n-k fusiones, la altura es la de la fusión (n-k-1)
    indice_fusion = n_puntos - k_grupos - 1
    if 0 <= indice_fusion < len(Z):
        altura_corte = Z[indice_fusion, 2]

# Crear el dendrograma
plt.figure(figsize=(15, 8))
dendrogram(Z, 
           labels=datos_complejos['ciudad'].values,  # Etiquetas de las ciudades
           leaf_rotation=90,  # Rotar las etiquetas 90 grados
           leaf_font_size=10)  # Tamaño de fuente

plt.title('Dendrograma: clustering jerárquico de ciudades\n(Altura = distancia entre grupos)', 
          fontsize=14, fontweight='bold')
plt.xlabel('Ciudades', fontsize=12, fontweight='bold')
plt.ylabel('Distancia', fontsize=12, fontweight='bold')
plt.axhline(y=altura_corte, color='r', linestyle='--', linewidth=2, label=f'Corte para 3 grupos (altura={altura_corte:.2f})')
plt.legend()
plt.tight_layout()
plt.show()

print("\nInterpretación del dendrograma:")
print("- Las ciudades más similares están unidas en la parte inferior")
print("- A mayor altura, más diferentes son los grupos que se unen")
print(f"- La línea roja muestra dónde 'cortar' para obtener 3 grupos (altura={altura_corte:.2f})")
print("- Puedes mover la línea horizontal para obtener diferentes números de grupos")
print("Las ciudades que se agrupan primero son las más similares")
````

## Limitaciones y consideraciones éticas

### Limitaciones del Clustering

Es importante entender qué no puede hacer el clustering y cuáles son sus limitaciones:

#### 1. **No explica causalidad**
- El clustering encuentra asociaciones, no causas.
- **Ejemplo**: Si Medellín y Bogotá están en el mismo grupo, no significa que una causa la otra.
- **Solución**: Usa clustering para generar hipótesis, no para probarlas.

#### 2. **Depende de las variables elegidas**
- Los grupos dependen completamente de qué variables incluyas.
- **Ejemplo**: Si solo usas variables económicas, encontrarás grupos económicos, no políticos.
- **Solución**: Basa la selección de variables en teoría política sólida.

#### 3. **Sensible a valores atípicos (outliers)**
- Un caso extremo puede distorsionar los grupos.
- **Ejemplo**: Incluir un país con valores extremos puede crear un grupo artificial.
- **Solución**: Identifica y analiza outliers por separado.

#### 4. **Asume que los grupos existen**
- El algoritmo encuentra grupos incluso si no existen.
- **Ejemplo**: Puede dividir un continuo en grupos arbitrarios.
- **Solución**: Valida con conocimiento teórico y métricas como el índice Silhouette.

#### 5. **No funciona bien con datos no lineales**
- K-means asume grupos esféricos o convexos.
- **Ejemplo**: Si los grupos tienen formas complejas, K-means puede fallar.
- **Solución**: Considera alternativas como clustering jerárquico o density-based.

#### 6. **Requiere decisiones subjetivas sobre el número de grupos**
- Aunque existen métricas, siempre hay un componente subjetivo.
- **Ejemplo**: Elegir entre tres o cuatro grupos depende de la pregunta de investigación.
- **Solución**: Usa métodos como el codo y Silhouette, junto con teoría.

---

### Consideraciones éticas en Clustering Político

Cuando aplicas clustering a datos políticos, hay consideraciones éticas clave:

#### 1. **Sesgos en los datos**
- Los datos pueden reflejar desigualdades estructurales.
- **Ejemplo**: Clasificar ciudades por “desarrollo” puede reforzar jerarquías históricas.
- **Solución**: Identifica y discute posibles sesgos.

#### 2. **Estigmatización de grupos**
- Nombrar grupos como “buenos” o “malos” puede ser problemático.
- **Ejemplo**: “Ciudades con bajo desempeño” puede estigmatizar territorios vulnerables.
- **Solución**: Presenta los grupos de forma neutral.

#### 3. **Uso político de los resultados**
- Los resultados pueden ser usados fuera de contexto.
- **Ejemplo**: “Medellín está en el grupo de alta criminalidad” puede ser exagerado.
- **Solución**: Acompaña siempre con contexto y limitaciones.

#### 4. **Privacidad y datos personales**
- En datos de votantes o ciudadanos, hay riesgos de privacidad.
- **Ejemplo**: Agrupar votantes puede revelar información sensible.
- **Solución**: Anonimiza datos y sigue normativas de protección.

#### 5. **Responsabilidad en la interpretación**
- La responsabilidad es tuya, no del algoritmo.
- **Ejemplo**: No atribuyas decisiones al algoritmo: tú elegiste las variables.
- **Solución**: Sé transparente en tus decisiones metodológicas.

## Resumen y guía de interpretación

### ¿Cuándo usar cada método?

| Método | Cuándo usarlo | Ventajas | Desventajas |
|--------|---------------|----------|-------------|
| **K-means** | Cuando tienes una idea aproximada del número de grupos | Rápido, fácil de interpretar | Requiere especificar k |
| **Jerárquico** | Cuando quieres explorar distintos niveles de agrupación | Visualización clara (dendrograma), flexible | Más lento con grandes volúmenes de datos |

---

### Checklist para tu análisis de clustering

Antes de aplicar clustering a tus datos:

- [ ] **¿Tus datos están limpios?** Sin valores faltantes ni errores.
- [ ] **¿Normalizaste las variables?** Importante si tienen escalas diferentes.
- [ ] **¿Elegiste el número adecuado de grupos?** Usa el método del codo y el índice silhouette.
- [ ] **¿Los grupos tienen sentido teórico?** Deben ser interpretables.
- [ ] **¿Validaste los resultados?** Contrasta con conocimiento previo del tema.

---

### Cómo comunicar los resultados

Al presentar tus resultados en un informe, trabajo académico o presentación:

1. **Describe el método usado**: “Aplicamos K-means con k = 3 grupos...”.

3. **Caracteriza cada grupo**: “El Grupo 1 presenta alta democracia y alto gasto público...”.
4. **Interpreta los resultados en su contexto**: “Estos grupos representan modelos distintos de política social...”.
5. **Menciona limitaciones**: “Los resultados dependen de las variables incluidas y pueden variar con otros datos...”.
